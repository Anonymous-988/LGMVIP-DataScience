<b># Task 4: Converting Image to Pencil Sketch Using Python</b>

```python
#import required libraries
import cv2
from google.colab.patches import cv2_imshow
```

```python
#load data from drive
data=cv2.imread("/content/drive/MyDrive/Data_Science/Datasets/ImageToPencilSketch/Image.jpg")

#resize data and store it as original image
original_image =cv2.resize(data, (700, 500))
cv2_imshow(original_image)
```

**Step 1: Convert Original RGB Image to Gray Image**

```python
#gray scale image
gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

**Step 2: Invert the Gray Image (negative image)**

```python
#gray image to be inverted
inverted_image = cv2.bitwise_not(gray_image)
cv2_imshow(inverted_image)
```

**Step 3: Use Gaussian Blur on Inverted Image**

```python
# Used to generate blurry inverted image
blurry_image = cv2.GaussianBlur(inverted_image, (21, 21),sigmaX=0, sigmaY=0)
cv2_imshow(blurry_image)
```

**Step 4: Mix Grayscale Image with Inverted Blurry Image**

```python
#create a function to divide() images
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256)

#call dodgeV2 function
pencil_sketch = dodgeV2(gray_image,blurry_image)
cv2_imshow(pencil_sketch)
```

```python
#Compare both Original and Final Image
cv2_imshow(original_image)
cv2_imshow(pencil_sketch)
```

**Conclusion:
Using above steps a plain RGB image can be converted to Pencil Sketch Image.**
